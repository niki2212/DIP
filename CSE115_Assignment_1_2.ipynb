```python
#List of different gamma values used for enhancement
gamma_list = [3.0, 4.0, 5.0, 6.0]
```

```python
#c is a constant
c = 1
```

**Power law Gamma Transformation**


It is used to change the brightness of the images
"normal_img" is the normalized image which is scaled between 0 and 1.

"i" takes values from the gamma_list which ranged values from 0.2 to 2.

"c" is the constant that is used to scale the image after the transformation, the pixel values will be from 0 to 255.

The "transformed_img" is then scaled back to the range of 0 to 255 by multiplying it with 255 and converting it to an unsigned 8-bit integer data type np.uint8.

```python
for i in gamma_list:
  normal_img = img/255.0
  transformed_img = c * np.power(normal_img, i)
  transformed_img = (transformed_img * 255).astype(np.uint8)

#Matplotlib to display the gamma-corrected image transformed_img
  plt.imshow(transformed_img, cmap='gray')
#setting the title
  plt.title('Transformed images  Gamma='+ str(i))
#displays the figure with the transformed image and the title
  plt.show()
```
