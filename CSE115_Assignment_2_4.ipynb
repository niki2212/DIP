```python
#reading the input image
input_img = cv2.imread('/content/drive/MyDrive/DIP/assignment 2/prob4.tif', cv2.IMREAD_GRAYSCALE)
plt.imshow(input_img, 'gray')
```

```python
#Applying median filter to the input image
filtered_img = cv2.medianBlur(input_img, 3)

#Displaying the input and output(filtered) image
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axs[0].imshow(input_img, cmap='gray')
axs[0].set_title('Input Noisy Image')
axs[1].imshow(filtered_img, cmap='gray')
axs[1].set_title('Filtered Image')
plt.show()
```

Median filter is applied for the given image which has salt and pepper noise. 
This filter helps to remove noise while preserving the edges and fine details in the image. It is a non-linear filtering technique. 

```python
#Applying min and max filter on the input image 

#dilate operation slides the kernel over the image and replaces each pixel 
#in the output image with the maximum value of the input pixels covered by the kernel
max_img = cv2.dilate(input_img, np.ones((3, 3), np.uint8))

#erode operation slides the kernel over the image and replaces each pixel 
#in the output image with the minimum value of the input pixels covered by the kernel
min_img = cv2.erode(input_img, np.ones((3, 3), np.uint8))

#Displaying the original image and the min and max filtered images
fig, axis = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
axis[0].imshow(input_img, cmap='gray')
axis[0].set_title('Original Image')
axis[1].imshow(max_img, cmap='gray')
axis[1].set_title('Max Filter')
axis[2].imshow(min_img, cmap='gray')
axis[2].set_title('Min Filter')
plt.show()
```

Here, we can see the input, max filter and min filter images. We can see that max filter enhanced the brighter refions of the image while suppressing the dark regions. It brings out the brighter regions of the image, but this also amplies the noise present in the input image. We can also see that in the min filter, the opposite is happening, it is removing all the small bright spots in the input image and enhacing all the dark regions in the image. But these filters caused loss of image details.
