```python
#reading the input image
input_img = cv2.imread('/content/drive/MyDrive/DIP/assignment 2/prob3.tif', cv2.IMREAD_GRAYSCALE)
plt.imshow(input_img, 'gray')
```

```python
#Defining the sizes to apply square averaging filters
m = [3, 5, 9, 25, 35]

#Apply the square averaging filters
fig, axis = plt.subplots(nrows=1, ncols = len(m)+1, figsize = (20, 5))

#showing the input image for reference
axis[0].imshow(input_img, cmap='gray')
axis[0].set_title('Original Image')

#Applying different sizes of m, for filter on the input image and showing all the images
for i, size in enumerate(m):
    kernel = np.ones((m,m), np.float32) / (m * m)
    filtered_img = cv2.filter2D(input_img, -1, kernel)
    axis[i+1].imshow(filtered_img, cmap='gray')
    axis[i+1].set_title(f'Filter Size(m) = {m}x{m}')
```

We can see that as the square averaging filter increases, the amount of smoothing in the filtered image increases. As the size of the kernel increases, it includes more neighboring pixels in the averaging process, which results in a more blurred image as seen in the above results. We can see that when the size of the kernel is 3x3 or 5x5, that is when it is small, the image preserves more of the input image's details, which also leads to noise in the output image. As the kernel size increases like 35x35, the noise is reduced but the sharpness is also reduced simultaneously.

```python
# Weighted averaging filter
kernel = np.array([[1,2,1], [2,4,2], [1,2,1]], np.float32) / 16

# Applying the weighted averaging filter on the input image
weighted_avg_img = cv2.filter2D(input_img, -1, kernel)

#Showing the original and weighted average filter to compare 
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axs[0].imshow(input_img, cmap='gray')
axs[0].set_title('Original Image')
axs[1].imshow(weighted_avg_img, cmap='gray')
axs[1].set_title('Weighted Average Filter')
plt.show()
```

Weighted averaging filter is applied on the input image, it helps reduce noise and improve overall appearance of the image. The output filter appears more smoother and has a better contrast.
