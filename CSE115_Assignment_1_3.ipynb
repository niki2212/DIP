```python
#importing libraries
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
#mounting google drive to access the images
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Path to the image file in the Google Drive directory
img = cv2.imread('/content/drive/MyDrive/DIP/prob3.tif')
```

```python
#Matplotlib function used to display an image
plt.imshow(img)
```

```python
#converting image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#intensity range for blood vessels
vessel_range = (145, 255)
```

```python
#creating a binary mask(each pixel is either 0 or 1)
mask = cv2.inRange(gray, vessel_range[0], vessel_range[1])
```

```python
#applyting the mask to our image
result = cv2.bitwise_and(img, img, mask=mask)
```

```python
#plotting both the original and result image to compare
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title('Result')
plt.axis('off')

plt.show()
```

In the above results, we can see that we have highlighted the blood vessels by using intensity slicing and checking for different values.

```python
#Path to the image file in the Google Drive directory
img = cv2.imread('/content/drive/MyDrive/DIP/prob3.tif', 0)
```

```python
#defining the range of threshold values to test
lower_range = 0
upper_range = 140
```

```python
#intensity level slicing for the given range
mask = cv2.inRange(img, lower_range, upper_range)
output = cv2.bitwise_and(img, img, mask=mask)
```

```python
#plotting both the original and result image to compare
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.title('Result Image')
plt.axis('off')

plt.show()
```

In the above images, I have highlighted the blood vessels and preserved the background intensity values.
